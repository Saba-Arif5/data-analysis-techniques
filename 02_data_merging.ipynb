**creating two dummy datasets to explore join(inner join, left join, outer join) and concat()**

```python
import pandas as pd
customers=pd.DataFrame({
    'customers_id':[1,2,3,4,6,7],
    'name':['Alice','Bob','Charlie','Diana','khadtija','Saba'],
    'region':['North','South','East','West','NorthWest','SouthWest'],
    
})
orders=pd.DataFrame({
    'order_id':[101,102,103,104,105,106],
   'customers_id':[1,5,2,3,4,8],
    'amount':[250,100,300,200,500,600],
    
})
inner_join=pd.merge(customers,orders,on='customers_id',how='inner')
display(inner_join)

```

```python
left_join=pd.merge(customers,orders,on='customers_id',how='left')
display(left_join)
```

```python
outer_join=pd.merge(customers,orders,on='customers_id',how='outer')
display(outer_join)
```

```python
extras=pd.DataFrame({
    'customers_id':[9,10],
    'name':['Bruno','Cocomo'],
    'region':['Canine','Feline']
    
    
})
concat_vertical=pd.concat([customers,extras], ignore_index=False)
display(concat_vertical)
```
